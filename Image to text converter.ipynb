```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path="coding.png"
reader=easyocr.Reader(['en'])
result=reader.readtext(image_path)
result
```

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][1])
text=result[0][1]
```

```python
img=cv2.imread(image_path)
img=cv2.rectangle(img,top_left,bottom_right,(255,255,0),2)
img=cv2.putText(img,text,top_left,cv2.FONT_HERSHEY_SIMPLEX,2,(255,0,0))
plt.imshow(img)
plt.show()
```

```python
for one in result:
    top_left = tuple(map(int, one[0][0]))
    bottom_right = tuple(map(int, one[0][2]))
    text = one[1]

    img = cv2.rectangle(img, top_left, bottom_right, (255, 255, 0), 2)
    img = cv2.putText(img, text, (top_left[0], top_left[1] - 10), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2)
plt.imshow(img)
plt.show()
```
